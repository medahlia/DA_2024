Імпортую бібліотеки:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Бібліотека для роботи з часовими рядами
import statsmodels.api as sm
import statsmodels.tsa.api as smt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

Зчитую датасет з туберкульозом:

```python
path = "cause_of_deaths.csv"
df = pd.read_csv(path)
df.head()
```

```python
countries = ["China", "India"]
filtered_df = df[df['Country/Territory'].isin(countries)]
filtered_df = filtered_df[filtered_df['Cause'] == 'Malaria']
filtered_df.head()
```

Часовий ряд для Китаю для малярії:

```python
plt.figure(figsize=(10, 6))


fig, ax = plt.subplots(figsize=(15, 10))

filtered_df.plot(ax=ax)
plt.title('Часова динаміка грипу в Україні та Польщі')
ax.grid()
plt.show()
```
